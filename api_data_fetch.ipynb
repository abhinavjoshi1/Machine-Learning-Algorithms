Conversion to Dataframe

```python
df = json_normalize(
    json_data,
    record_path=['bookmakers', 'markets', 'outcomes'],
    meta=[
        'id',
        'sport_key',
        'sport_title',
        'commence_time',
        'home_team',
        'away_team',
        ['bookmakers', 'key'],
        ['bookmakers', 'title'],
        ['bookmakers', 'last_update'],
        ['bookmakers', 'markets', 'key']
    ],
    errors='ignore'
)
print(df)
```
